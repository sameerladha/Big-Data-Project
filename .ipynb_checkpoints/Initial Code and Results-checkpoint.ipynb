```python
import pandas as pd
from patsy import dmatrices
import numpy as np
from sklearn.model_selection import train_test_split
import statsmodels.discrete.discrete_model as sm
from sklearn import preprocessing

df = pd.read_csv('wpmhdata.csv', sep=',')
df = df.iloc[: , 1:-5]
```

**Recoding Values in order to accurately conduct Naive Bayes and Decision Tree algorithms**

This section of the code will recode and prepare the data in the python environemnt to be analysed by the required algorithms

```python
#This variable lists all the categorical variables 

cat_cols=["sex","sex","tech_nontech","no_Employees","benefits","employer_resources","anonymity_protected","leave","discussing_mh_negative","discussing_ph_negative","comfortable_coworker", "comfortable_supervisor","physical_serious", "consequences_discussed","productivity_affected","amount_productivity_affected","identifing_career","coworker_negative_view",
          "share_ff","bad_response_experience","see_others_reveal","current_disorder","diagnosed","treatment","interfere_treated","interferes_not_treated"]
```

```python
df_onehot=df.copy()

#convert only categorical variables/features to dummy/one-hot features
df_onehot = pd.get_dummies(df, columns=cat_cols, prefix = cat_cols)
df_onehot = df_onehot.iloc[:,4:]

df_onehot.info()
```

```python
from sklearn.model_selection import train_test_split

class_col_name="productivity_affected_Yes"

one_hot_feature_names=df_onehot.columns[df_onehot.columns != class_col_name]

# Split dataset into training set and test set
X = list(set(list(df_onehot)) - set([class_col_name]))

X_train, X_test, y_train, y_test = train_test_split(df_onehot[X], df_onehot[class_col_name], test_size=0.3,random_state=0) # 70% training and 30% test
```

**Naive Bayes Algorithm**

```python
from sklearn.naive_bayes import MultinomialNB

nb = MultinomialNB()

#Train the model using the training sets
nb.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = nb.predict(X_test)
```

```python
#This section will check the number of features and records to ensure they match with our original data set and that all records are analyzed appropriately

print("Number of features used ",nb.n_features_in_)
print("Classes ",nb.classes_)
print("Number of records for classes ",nb.class_count_)
```

**Evaluation of Naive Bayes**

In this section, I will evaluate the Naive Bayes model using a confusion matrix. This will allow me to determine if the data from this model is useful and provides relevant insights.

```python
from sklearn.metrics import confusion_matrix

cf=confusion_matrix(y_test, y_pred)
print ("Confusion Matrix")
print(cf)
tn, fp, fn, tp=cf.ravel()
print ("TP: ", tp,", FP: ", fp,", TN: ", tn,", FN:", fn)
```

```python
from sklearn.metrics import classification_report
from sklearn import metrics

print(classification_report(y_test, y_pred))
```

**Decision Tree**

```python
from sklearn import tree
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)

import graphviz
dot_data = tree.export_graphviz(clf, feature_names=one_hot_feature_names)

graph = graphviz.Source(dot_data, format="pdf") 
graph
```

**Notes on output**

The for loops below allows us to quickly see the gini importance of notable features

```python
rangelen = clf.feature_importances_
allgini = []
for i in range(len(rangelen)):
    hold = []
    if clf.feature_importances_[i] > 0:
        hold.append(clf.feature_names_in_[i])
        hold.append(clf.feature_importances_[i])
        allgini.append(hold)
allgini = sorted(allgini,key=lambda x: x[1])
print ("The Class used in this tree is :",class_col_name)
print ()
for next in allgini:
    print (next)
```

**Evaluation of Decision Tree**

```python
# Perform prediction on the test set
y_pred = clf.predict(X_test)

# Get classification report
from sklearn.metrics import classification_report
from sklearn import metrics

print(classification_report(y_test, y_pred))
```

**KNN**

This section will utilize the KNN algorithim from sklearn to analyze the data.

```python
from sklearn.neighbors import KNeighborsClassifier  
x_trainKNN, x_testKNN, y_trainKNN, y_testKNN = train_test_split(df_onehot[X], df_onehot['productivity_affected_Yes'],test_size=0.30, random_state=25)

classifier = KNeighborsClassifier(n_neighbors=3) 
classifier.fit(x_trainKNN, y_trainKNN)
y_predKNN = classifier.predict(x_testKNN)
classifier.score(x_trainKNN, y_trainKNN)
```

```python
print ('KNN Confusion Matrix')
predKNN = y_predKNN > 0.5
actualKNN = y_testKNN > 0.5
KNNcm = confusion_matrix(actualKNN, predKNN)
print(KNNcm)
```
